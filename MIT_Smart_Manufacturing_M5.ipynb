**Compare os resultados das perguntas 1 e 2:**


* **Que mudança resulta em uma resposta mais rápida?**

O tempo de resposta mais rápido é encontrado na mudança de diâmetro alvo de um valor alto para um valor baixo.

* **Como a mudança nos tempos de resposta pode ser explicada?**

A mudança nos tempos de resposta pode ser explicada considerando os princípios do funcionamento do FrED. Quando se deseja reduzir o diâmetro da fibra, a velocidade angular é aumentada, de modo que, o tempo de resposta é menor pois o material está sendo empurrado na mesma direção do fluxo de massa. Por outro lado, quando se deseja aumentar o diâmetro da fibra, a velocidade angular é reduzida e temos um tempo maior de ajuste para que seja controlado o fluxo de massa para dentro da extrusora.
