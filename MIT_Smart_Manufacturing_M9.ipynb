```python
# Importar as Bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from numpy.fft import fft, ifft

plt.style.use('seaborn-pastel')
```

**FrED para produção da fibra de 0.3mm de diâmetro**

```python
# Carregar os dados e visualização prévia
df_SCF_3 = pd.read_excel('MITPE_Smart_Manufacturing_Module9_Experiment_ClosedLoop_0.3mm_(1)__[PT].xlsx')
print('O dataframe possui',len(df_SCF_3),'registros organizados em',len(df_SCF_3.columns),'colunas.')
print('')
df_SCF_3.head()
```

```python
# Representar graficamente as variáveis em função do tempo:
print('Nos gráficos a seguir podemos observar o comportamento das variáveis em função do tempo:')
print('')

plt.figure(figsize=(25, 20))

plt.subplot(3, 1, 1)
plt.hist(df_SCF_3['Diâmetro (mm) - Diâmetro medido'])
plt.title('Histograma')
plt.xlabel('Dinâmetro (mm)')
plt.ylabel('Contagem')

plt.subplot(3, 1, 2)
plt.plot(df_SCF_3[['Tempo - Diâmetro medido']],df_SCF_3[['Diâmetro (mm) - Diâmetro medido']],marker='.',linewidth=2)
plt.plot(df_SCF_3[['Tempo - Diâmetro determinado']],df_SCF_3[['Diâmetro (mm) - Diâmetro determinado']],color = 'pink',linewidth=3)
plt.title('Gráfico de Série Temporal')
plt.xlabel('Tempo (s)')
plt.ylabel('Diâmetro (mm)')

plt.subplot(3, 1, 3)
plt.plot(df_SCF_3[['Tempo - Diâmetro medido']],df_SCF_3[['Diâmetro (mm) - Diâmetro medido']].rolling(50).mean(),marker='.',linewidth=2)
plt.plot(df_SCF_3[['Tempo - Diâmetro determinado']],df_SCF_3[['Diâmetro (mm) - Diâmetro determinado']],color = 'pink',linewidth=3)
plt.title('Gráfico de Série Temporal')
plt.xlabel('Tempo (s)')
plt.ylabel('Diâmetro (mm)')

plt.show()
```

**Comentários:**

* As medições em tempo real, para um diâmetro da fibra de 0.3mm não são normalmente distribuídas. Existem flutuações periódicas ao longo do limite central e podem estar relacionadas a fatores externos como um novo material, excentricidades no eixo, entre outros. Logo, este processo não está sob controle estatístico.
* O histograma também indica que o processo não é normalmente distribuído.
* Aplicando a média móvel a cada 50 pontos, observa-se que o processo aparente uma distribuição normal e parece estar sob controle estatísitico. 
* O gráfico indica que as variações não são aleatórias.

**FrED para produção da fibra de 0.4mm de diâmetro**

```python
# Carregar os dados e visualização prévia
df_SCF_4 = pd.read_excel('MITPE_Smart_Manufacturing_Module9_Experiment_ClosedLoop_0.4mm_(2)__[PT].xlsx')
print('O dataframe possui',len(df_SCF_4),'registros organizados em',len(df_SCF_4.columns),'colunas.')
print('')
df_SCF_4.head()
```

```python
# Representar graficamente as variáveis em função do tempo:
print('Nos gráficos a seguir podemos observar o comportamento das variáveis em função do tempo:')
print('')

plt.figure(figsize=(25, 20))

plt.subplot(3, 1, 1)
plt.hist(df_SCF_4['Diâmetro (mm) - Diâmetro medido'])
plt.title('Histograma')
plt.xlabel('Dinâmetro (mm)')
plt.ylabel('Contagem')

plt.subplot(3, 1, 2)
plt.plot(df_SCF_4[['Tempo - Diâmetro medido']],df_SCF_4[['Diâmetro (mm) - Diâmetro medido']],marker='.',linewidth=2)
plt.plot(df_SCF_4[['Tempo - Diâmetro determinado']],df_SCF_4[['Diâmetro (mm) - Diâmetro determinado']],color = 'pink',linewidth=3)
plt.title('Gráfico de Série Temporal')
plt.xlabel('Tempo (s)')
plt.ylabel('Diâmetro (mm)')

plt.subplot(3, 1, 3)
plt.plot(df_SCF_4[['Tempo - Diâmetro medido']],df_SCF_4[['Diâmetro (mm) - Diâmetro medido']].rolling(125).mean(),marker='.',linewidth=2)
plt.plot(df_SCF_4[['Tempo - Diâmetro determinado']],df_SCF_4[['Diâmetro (mm) - Diâmetro determinado']],color = 'pink',linewidth=3)
plt.title('Gráfico de Série Temporal')
plt.xlabel('Tempo (s)')
plt.ylabel('Diâmetro (mm)')

plt.show()
```

**Comentários:**

* As medições em tempo real, para um diâmetro da fibra de 0.4mm não são normalmente distribuídas. Existem mais flutuações na descida do que na subida, acima do limite central, e podem estar associadas a fatores externos como aumento da velocidade angular do FrED, recalibração da temperatura, entre outros. Logo, este processo não está sob controle estatístico.
* O histograma também indica que a variação está se tornando menor e se movendo para fora dos limites de específicação.
* Aplicando a média móvel a cada 125 pontos, observa-se que o processo aparente uma distribuição normal e parece estar sob controle estatísitico. 
* O gráfico indica que as variações não são aleatórias.
