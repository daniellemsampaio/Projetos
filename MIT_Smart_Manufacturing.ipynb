```python
# Importar as Bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-pastel')
```

```python
# Carregar os dados e visualização prévia
df = pd.read_csv('MITPE_Smart_Manufacturing_Module1_Dataset_Time_based_visualization__[PT]-1.csv',engine='python')
df = df.rename(columns={'Rota��es/segundo bobina - Velocidade medida': 'Rotacoes/segundo bobina - Velocidade medida'})
df = df.rename(columns={'Tempo - Di�metro medido': 'Tempo - Diametro medido'})
df = df.rename(columns={'Di�metro (mm) - Di�metro medido': 'Diametro (mm) - Diametro medido'})
df = df.rename(columns={'Rota��es/segundo bobina - Velocidade-alvo': 'Rotacoes/segundo bobina - Velocidade-alvo'})
df = df.rename(columns={'Tempo - Di�metro previsto': 'Tempo - Diametro previsto'})
df = df.rename(columns={'Di�metro (mm) - Di�metro previsto': 'Diametro (mm) - Diametro previsto'})
df = df.rename(columns={'Di�metro (mm) - Temperatura': 'Diametro (mm) - Temperatura'})
print('O dataframe possui',len(df),'registros organizados em',len(df.columns),'colunas.')
print('')
df.head()
```

```python
# Verificar os tipos de dados
print('Na tabela abaixo podemos observar os tipos de dados do dataframe:')
print('')
pd.DataFrame({'tipos':df.dtypes, 'missing':df.isna().sum()})
```

```python
# Representar graficamente as variáveis em função do tempo
print('Nos gráficos a seguir podemos observar o comportamento das variáveis em função do tempo:')
print('')

plt.figure(figsize=(20, 14))

plt.subplot(3, 2, 1)
plt.scatter(df[['Tempo - Velocidade medida']],df[['Rotacoes/segundo bobina - Velocidade medida']],marker='.')
plt.plot(df[['Tempo - Velocidade medida']],df[['Rotacoes/segundo bobina - Velocidade medida']],linewidth=2)
plt.plot(df[['Tempo - Velocidade-alvo']],df[['Rotacoes/segundo bobina - Velocidade-alvo']],color = 'pink',linewidth=3)
plt.title('Gráfico de Dispersão')
plt.xlabel('Tempo (s)')
plt.ylabel('Velocidade (rps)')

plt.subplot(3, 2, 2)
plt.hist(df['Rotacoes/segundo bobina - Velocidade medida'])
plt.title('Histograma')
plt.xlabel('Velocidade Medida (rps)')
plt.ylabel('Contagem')

plt.subplot(3, 2, 3)
plt.scatter(df[['Tempo - Diametro medido']],df[['Diametro (mm) - Diametro medido']],marker='.')
plt.plot(df[['Tempo - Diametro medido']],df[['Diametro (mm) - Diametro medido']],linewidth=2)
plt.plot(df[['Tempo - Diametro previsto']],df[['Diametro (mm) - Diametro previsto']],color = 'pink',linewidth=3)
plt.xlabel('Tempo (s)')
plt.ylabel('Diametro medido (mm)')

plt.subplot(3, 2, 4)
plt.hist(df['Diametro (mm) - Diametro medido'])
plt.xlabel('Diametro medido (mm)')
plt.ylabel('Contagem')

plt.subplot(3, 2, 5)
plt.scatter(df[['Tempo - Diametro medido']],df[['Diametro (mm) - Temperatura']],marker='.')
plt.plot(df[['Tempo - Diametro medido']],df[['Diametro (mm) - Temperatura']],linewidth=2)
plt.xlabel('Tempo (s)')
plt.ylabel('Temperatura (°C)')

plt.subplot(3, 2, 6)
plt.hist(df['Diametro (mm) - Temperatura'])
plt.xlabel('Temperatura (°C)')
plt.ylabel('Contagem')

plt.show()
```

```python
# Representar graficamente a correlação entre duas variáveis
print('Nos gráficos a seguir podemos observar a relação entre a velocidade e a temperatura com o Diâmetro:')
print('')

plt.rcParams['figure.figsize'] = [20, 5]

fig,ax = plt.subplots()
ax.plot(df[['Tempo - Velocidade medida']], df[['Rotacoes/segundo bobina - Velocidade medida']],color='lightskyblue')
ax.set_xlabel('Tempo (s)')
ax.set_ylabel('Velocidade (rps)',color='lightskyblue')
ax2=ax.twinx()
ax2.plot(df[['Tempo - Diametro medido']], df[['Diametro (mm) - Diametro medido']],color='mediumpurple')
ax2.set_ylabel('Diametro (mm)',color='mediumpurple')
plt.title('Grafico do Diametro vs. Velocidade')
plt.show()

fig,ax = plt.subplots()
ax.plot(df[['Tempo - Diametro medido']],df[['Diametro (mm) - Temperatura']],color='lightsalmon')
ax.set_xlabel('Tempo (s)')
ax.set_ylabel('Temperatura (°C)',color='lightsalmon')
ax2=ax.twinx()
ax2.plot(df[['Tempo - Diametro medido']], df[['Diametro (mm) - Diametro medido']],color='mediumpurple')
ax2.set_ylabel('Diametro (mm)',color='mediumpurple')
plt.title('Grafico do Diametro vs. Temperatura')
plt.show()
```

```python
# Analisar a correlação entre as variáveis:
print('No gráfico de calor a seguir, podemos observar a correlação entre as variáveis:')
print('')
ax = sns.heatmap(df[['Rotacoes/segundo bobina - Velocidade medida','Diametro (mm) - Diametro medido','Diametro (mm) - Temperatura']].corr(),  linewidths=1, annot=True, cmap='Pastel2')
```
